## 라인그래프

```python
# 엑셀 파일 읽어오기
df = pd.read_excel('./data/시도_별_이동자수.xlsx', header =0)
print(df.head())
```

```python
# 엑셀에서 셀병합이 있으면 NaN으로 처리 -전국
#NaN 데이터를 앞에 데이터로 채우기
df = df.fillna(method = 'ffill')
print(df.head())
```

```python
# 전출지별 = 서울특별시, 전입지별 =! 서울인 데이터만 추출!
# 조건 생성
mask = (df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')
df_seoul = df[mask]
print(df_seoul.head())
```

```python
# 데이터가 없는 1970~1985년대 데이터 삭제하기
li =[]
for i in range(1970, 1986) :
    li.append(str(i))
    
print(li)

df_seoul.drop([li], axis = 1 , inplace = True)
print(df_seoul.head())
```

```python
# df_seoul.drop(['전출지별'], axis = 1, inplace = True)
# print(df_seoul.head())
# df_seoul.rename({'전입지별':'전입지'}, axis = 1, inplace = True)
df_seoul.set_index('전입지', inplace= True)
print(df_seoul.head())
```

```python
#전라남도만 추출하기
sr_one = df_seoul.loc['전라남도']
print(sr_one)
```

```python
#sr_one 데이터를 가지고 라인 그래프 그릭;
import matplotlib.pyplot as plt
plt.plot(sr_one.index, sr_one.values)
plt.show()
```

## seaborn

```python
#회귀에서 사용하는 데이터 셋 가져오기
import seaborn as sns
tips = sns.load_dataset('tips')
print(tips.head())
```

```python
#산점도와 회귀식을 출력
plt.figure(figsize=(8,6)) # 캔버스 크기 설정

#hue에 카테고리 별로 구분해서 그려줌!!
sns.lmplot(x = 'total_bill', y = 'tip', hue = 'smoker',data= tips)
plt.show()
```

```python
# haetmap 으로 그림 그리기
flights = sns.load_dataset('flights')
# print(flights)
# print(flights.pivot(columns=('month', 'year','passengers')))

# index를 month로 column 에 year를 passengers 의 핪계를 구하는 피벗을 생성
data = flights.pivot(index = 'month', columns = 'year', values = 'passengers')
# print(data)
# 포멧은 정수로
sns.heatmap(data, annot = True, fmt = 'd', annot_kws = {'size': 7})
plt.show()
```
